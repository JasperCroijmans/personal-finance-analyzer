```python
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Laad Rabobank CSV (speciaal formaat)
df = pd.read_csv('../data/raw/transactions.csv', 
                 sep=',',
                 encoding='latin-1',
                 quotechar='"',
                 skipinitialspace=True)

# Bekijk eerste rijen
print("=" * 60)
print("FIRST 10 TRANSACTIONS")
print("=" * 60)
print(df.head(10))

# Basis info
print("\n" + "=" * 60)
print("DATASET INFO")
print("=" * 60)
print(f"Total transactions: {len(df):,}")
print(f"\nColumns ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")

print(f"\nData types:")
print(df.dtypes)

# Check for missing values
print(f"\n" + "=" * 60)
print("MISSING VALUES")
print("=" * 60)
missing = df.isnull().sum()
if missing.sum() == 0:
    print("✅ No missing values!")
else:
    print(missing[missing > 0])

# Basic statistics
print(f"\n" + "=" * 60)
print("SHAPE")
print("=" * 60)
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")

# Show some sample amounts
print(f"\n" + "=" * 60)
print("SAMPLE BEDRAGEN (AMOUNTS)")
print("=" * 60)
if 'Bedrag' in df.columns:
    print(df['Bedrag'].head(10))
```
